```python
# Melakukan Import Library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
```

```python
data = {
    "X": [5, 4, 6, 5, 5, 5, 6, 6, 2, 7, 7],  # Usia Mobil (tahun)
    "Y": [85, 103, 70, 82, 89, 98, 66, 95, 169, 70, 46],  # Harga Mobil ($100)
}
```

```python
# Konversi ke DataFrame
dataset = pd.DataFrame(data)
```

```python
# Menampilkan kunci kolom pada dataset
print("Kolom dalam dataset:", dataset.keys())
```

```python
# Menampilkan dimensi dataset
print("Dimensi dataset:", dataset.shape)
```

```python
# Menampilkan isi sebagian dataset
dataku = pd.DataFrame(dataset)
print("Isi sebagian dataset:")
print(dataku.head(15))
```

```python
# Memisahkan variabel input (X) dan output (Y)
X = dataset["X"].values.reshape(-1, 1)
Y = dataset["Y"].values
```

```python
# Split dataset menjadi training set dan testing set
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
# Fitting Simple Linear Regression pada training set
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, Y_train)
model = LinearRegression()
model.fit(X, Y)
```

```python
# Menampilkan persamaan regresi
a = model.coef_[0]  # Koefisien regresi
b = model.intercept_  # Intercept regresi
print(f"Persamaan Regresi: Y = {a:.2f}X + {b:.2f}")
```

```python
# Memprediksi hasil Test-set
Y_pred = regressor.predict(X_test)
```

```python
# Visualisasi Data
plt.figure(figsize=(10,8))
plt.scatter(X, Y, color='blue', label='Data Aktual')
plt.xlabel('Usia Mobil (Tahun)')
plt.ylabel('Harga Mobil ($100)')
plt.title("Grafik Usia Mobil & Harga Mobil")
plt.show()
```

```python
# Visualisasi data dan garis regresi
plt.figure(figsize=(10,8))
plt.scatter(X, Y, color='blue', label='Data Aktual')
plt.plot(X, model.predict(X), color='red', label='Garis Regresi')
plt.xlabel('Usia Mobil (Tahun)')
plt.ylabel('Harga Mobil ($100)')
plt.title("Grafik Usia Mobil & Harga Mobil (Garis Regresi)")
plt.legend()
plt.show()
```

```python
# Visualisasi Hasil Prediksi Pada Training-Set
# Ukuran Plot
plt.figure(figsize=(10,8))

# Biru adalah data observasi
plt.scatter(X_train, Y_train, color = 'blue')

# Garis merah adalah hasil prediksi dari machine leraning
plt.plot(X_train, regressor.predict(X_train), color = 'red')

# Memberi Judul dan Label
plt.title('Grafik Usia Mobil & Harga Mobil (Training Set)')
plt.xlabel('Usia Mobil (Tahun)')
plt.ylabel('Harga Mobil ($100)')
```

```python
# Visualisasi Hasil Prediksi Pada Training-Set
# Ukuran Plot
plt.figure(figsize=(10,8))

# Biru adalah data observasi
plt.scatter(X_test, Y_test, color = 'blue')

# Garis merah adalah hasil prediksi dari machine leraning
plt.plot(X_test, regressor.predict(X_test), color = 'red')

# Memberi Judul dan Label
plt.title('Grafik Usia Mobil & Harga Mobil (Training Set)')
plt.xlabel('Usia Mobil (Tahun)')
plt.ylabel('Harga Mobil ($100)')

# Menampilkan Plot
plt.show()
```
