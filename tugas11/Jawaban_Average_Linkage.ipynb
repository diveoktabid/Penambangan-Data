```python
# Import library
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np
```

```python
# Data skor kedisiplinan siswa
x = np.array([
    [2, 4, 4, 3],  # Siswa 1
    [4, 3, 5, 5],  # Siswa 2
    [3, 2, 3, 5],  # Siswa 3
    [1, 5, 4, 2],  # Siswa 4
    [5, 2, 1, 3]   # Siswa 5
])
```

```python
# Lakukan hierarchical clustering (Average Linkage)
z_average = linkage(x, method='average', metric='cityblock')
```

```python
# Plot dendrogram
plt.figure(figsize=(10, 5))
plt.title('Dendrogram Hierarchical Klastering (Average Linkage)')
plt.xlabel('Siswa')
plt.ylabel('Jarak')
dendrogram(z_average, labels=[1, 2, 3, 4, 5])
plt.show()
```
