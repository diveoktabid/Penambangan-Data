```python
#Import librarynya terlebih dahulu
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np
```

```python
# buat array 2D ke dalam varuabel x, Data ini yang akan kita klaster.
x = np.array ([[1,2],[2.5,4.5],[2,2],[4,1.5],[4,2.5]])
```

```python
#lakuakn hierachical klastering
z = linkage(x)
```

```python
#plot dendogram
plt.figure(figsize=(25, 10))
plt.title('Dendrogram Hierarchical Klastering')
plt.xlabel('Data')
plt.ylabel('Jarak')
dendrogram(z)
plt.show()

```
